```python
#Loading data into Pandas
import pandas as pd

df = pd.read_csv(r'C:\Users\dell\Desktop\Data Analyst\Python\csv files\pokemon_data.csv')

df['HP']
```

```python
#Reading Data in Pandas
df.columns
```

```python
#Sorting/Describing Data
df.sort_values(['Type 1', 'HP'], ascending=[1,0])

df
```

```python
#Making changes to the data
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]

df.head(5)
```

```python
#Saving our Data 
df.to_csv(r'C:\Users\dell\Desktop\Data Analyst\Python\csv files\modified.csv', index=False)
```

```python
#Filtering Data
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

new_df.reset_index(drop=True, inplace=True)

new_df
```

```python
new_df.to_csv(r'C:\Users\dell\Desktop\Data Analyst\Python\csv files\filtered.csv')
```

```python
#Conditional Changes
df.loc[df['Total'] > 500, ['Generation','Legendary']] = ['Test 1', 'Test 2']

df
```

```python
#Aggregate Statistics (Groupby)
df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

```python

new_df = pd.DataFrame(columns=df.columns)


new_df
```

```python
#Working with large amounts of data

for df in pd.read_csv(r'C:\Users\dell\Desktop\Data Analyst\Python\csv files\modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    results = results.sort_index()
    
    new_df = pd.concat([new_df, results], sort=True)
```

```python
type_dfs = []  # List to store DataFrames for each type

for df in pd.read_csv(r'C:\Users\dell\Desktop\Data Analyst\Python\csv files\modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    type_dfs.append(results)

# Concatenating DataFrames for each type to display them consecutively
result_concatenated = pd.concat(type_dfs)
result_concatenated
```
